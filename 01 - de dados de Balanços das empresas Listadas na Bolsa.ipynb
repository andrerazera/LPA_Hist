```python
import pandas as pd
import wget
# from google.colab.data_table import DataTable
from zipfile import ZipFile
```

###Acessando a base de dados e criando arquivos históricos

```python
# url_base = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/' #link para ITR


url_base = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/' #link DFP

```

* Criando uma lista com o nome de todos os arquivos

```python
range_anos = range(2011,2022)

arquivos_zip = []
for ano in range_anos:
  # arquivos_zip.append(f'itr_cia_aberta_{ano}.zip') #para ITR
  arquivos_zip.append(f'dfp_cia_aberta_{ano}.zip') #para DFP

arquivos_zip
```

* fazendo o download de todos os arquivos

```python
for arq in arquivos_zip:
  wget.download(url_base+arq)
```

* extraindo os arquivos zip

```python
for arq in arquivos_zip:
  ZipFile(arq, 'r').extractall('CVM')
```

* criando arquivos históricos

```python
!mkdir DADOS

nomes = ['BPA_con',
         'BPA_ind',
         'BPP_con',
         'BPP_ind',
         'DFC_MD_con',
         'DFC_MD_ind',
         'DFC_MI_con',
         'DFC_MI_ind',
         'DMPL_con',
         'DMPL_ind',
         'DRE_con',
         'DRE_ind',
         'DVA_con',
         'DVA_ind']

for nome in nomes:
  arquivo = pd.DataFrame()
  for ano in range(2011,2023):
    try:
      arquivo = pd.concat([arquivo, pd.read_csv(f'CVM/dfp_cia_aberta_{nome}_{ano}.csv', sep=';', encoding='ISO-8859-1')])
    except:
      pass

  arquivo.to_csv(f'DADOS/dfp_cia_aberta_{nome}_2011-2021', index=False)
```
